### Use IQR to identify potential outliers

```python
for i in numeric_cols:
    stat = train_data_df[i].describe()
    print(stat)
    IQR = stat['75%'] - stat['25%']
    upper = stat['75%'] + 1.5 * IQR
    lower = stat['25%'] - 1.5 * IQR
    print('The upper and lower bounds for suspected outliers are {} and {}.'.format(upper, lower))
```

```python
def encode_label(df, col):
    #encode the categories using age for each category to replace label 
    cat_dict = {}
    cats = train_data_df[col].cat.categories.tolist()
    for cat in cats:
        cat_dict[cat] = train_data_df[train_data_df[col] == cat]['Age'].mean()
        print(cat_dict)
    train_data_df[col] = train_data_df[col].map(cat_dict)    
```

```python
for col in categorical_cols:
    encode_label(train_data_df, col)
```

```python
# Correlations between selected features and response
# jobId is discarded because it is unique for individual
fig = plt.figure(figsize = (12,10))
features = ['Daily_Time_Spent_on_Site', 'Age', 'Area_Income',
       'Daily_Internet_Usage', 'Ad_Topic_Line', 'City', 'Male', 'Country']
sns.heatmap(train_data_df[features + ['Clicked_on_Ad']].corr(), cmap='Blues', annot = True)
plt.xticks(rotation=45)
plt.show()
```

We see that Ad_Topic_Line and City are strongly correlated with Age.
This is because the mean of Age is used to label the categorical data.
